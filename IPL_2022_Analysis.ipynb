**MORE NUMBER OF MATCHES WON BY THE TEAM IN THE IPL 2022 ANALYSIS**

```python
#More No of matches won in the IPL 2022
plt.title("No of matches won in IPL 2022",fontweight = "bold")
plt.xlabel("Teams")
plt.ylabel("No.of matches won")
color = ["#348498"]
plt.hist(df['match_winner'],color=color);
plt.rcParams["figure.figsize"] = (16,8)
sns.set_style("darkgrid")
```

**OBSERVATION**

*   Gujarat Lions won the IPL Trophy with leading the tournament by winning the twelve matches.
*   Rajasthan Royals won the RUNNER in the IPL 2022 by winning the almost 10 matches.





```python
#No of matches won by chasing or defending

df['won_by'] = df['won_by'].map({"Wickets":"Chasing","Runs":"Defending"})
won_by = df['won_by'].value_counts()
header = won_by.index
values = won_by.values
plt.title("No of matches won by chasing or defending",fontweight = "bold")
color = ["#F71E35","#1794A5"]
plt.pie(values,colors = color,startangle = 45,labels = header,autopct='%.0f%%');
plt.legend();
```

**OBSERVATION**

*   Most of the teams in Indian Premier League are won half of the match while chasing the target and remaining half of the match won while defending the target in IPL 2022.



```python
#Toss decision

toss = df['toss_decision'].value_counts()
header = toss.index
values = toss.values
plt.title("Toss decision",fontweight = "bold")
color = ["#08ffc8","#204969"]
plt.pie(values,colors = color,labels = header,autopct='%.0f%%');
plt.legend();
```

**OBSERVATION**

*   Most of the team captains choose to field first after winning the toss. So far, in 59 matches captains have chosen to field first, and 15 games, the captains have chosen to bat first.



```python
#Venues of the most of the matches

venue = df['venue'].value_counts()
header = venue.index
values = venue.values
plt.title("Match Venue in IPL 2022",fontweight = "bold")
color = ["#515bd4","#8134af","#dd2a7b","#feda77","#f2f4f6","#1ee3cf"]
plt.pie(values,colors = color,labels = header,autopct='%.0f%%');
plt.legend(bbox_to_anchor=(1,0));
```

**OBSERVATION**

*   The analysis shows the most of the matches played in the Wankhede Stadium and Dr DY Patil Sports Academy in Mumbai.



```python
#Top scorer in IPL 2022
plt.hist(df['top_scorer'],bins = 37,color = 'orange')
plt.xticks(rotation =90);
plt.xlabel("top scorer")
plt.ylabel("count")
plt.title("Top Scrores in IPL 2022",fontweight = "bold");
```

**OBSERVATION**

*   Jos Buttler has been a top scorer in 5 matches. He was played for Rajasthan Royals team.



```python
#Best Bowler
plt.hist(df['best_bowling'],bins = 37,color = 'purple')
plt.xticks(rotation =90);
plt.xlabel("Bowler Name")
plt.ylabel("count")
plt.title("Best Bowler in IPL 2022",fontweight = "bold");
```

**OBSERVATION**

*    Yuzvendra Chahal having the best bowling figures in four matches in IPL 2022.



```python
#Most player of the match
plt.hist(df['player_of_the_match'],bins=56,color = "#0000A1");
plt.xticks(rotation=90);
plt.xlabel("Player Name")
plt.ylabel("count")
plt.title("Most Player of the Match Awards",fontweight = "bold");
```

**OBSERVATION**

*   Kuldeep Yadav is leading in the list of players of the match awards with four matches. 
*   It is a great tournament for Kuldeep Yadav this year.



```python
#Most wcket falling in ground wise
figure = go.Figure()
figure.add_trace(go.Bar(
    x=df["venue"],
    y=df["first_ings_wkts"],
    name='First Innings Wickets',
    marker_color='#400D12'
))
figure.add_trace(go.Bar(
    x=df["venue"],
    y=df["second_ings_wkts"],
    name='Second Innings Wickets',
    marker_color='#4FD5D6'
))
figure.update_layout(barmode='group', xaxis_tickangle=50,title="Most wicket's falling in the ground")
```

**OBSERVATION**

*   So in the Wankhede Stadium in Mumbai and MCA Stadium in Pune, most wickets fall while chasing the target.
*   And in the other two stadiums, most wickets fall while setting the target.
